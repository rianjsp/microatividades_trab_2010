# Microatividade 5
- InformaÃ§Ãµes gerais sobre as colunas, linhas e dados

```python
dados = pd.read_csv('./archive/earthquake_data_tsunami.csv', sep=',')

#Imprimir infos gerais do conjunto
print('ðŸ”¹ InformaÃ§Ãµes gerais do DataFrame: ')
print(dados.info())

#Total de linhas e colunas
linhas, colunas = dados.shape
print(f'\n ðŸ”¹ Total de Linhas: {linhas}')
print(f'ðŸ”¹ Total de colunas: {colunas}')

#Quantidade de dados nulos por coluna
print('\n ðŸ”¹ Quntidade de dados nulos por coluna: ')
print(dados.isnull().sum())

#Tipos de dados por coluna
print('\n ðŸ”¹Tipo de dados por coluna: ')
print(dados.dtypes)

#Memoria utilizada pelo DF
print('\n ðŸ”¹Memoria utilizada pelo conjunto de dados: ')
print(dados.memory_usage(deep=True).sum(), 'bytes')
```
